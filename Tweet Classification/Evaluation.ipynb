# Model Evaluation

```python
import evaluator
import numpy as np
import pandas as pd
```

Note: Evaluator class located in "evaluator.py" is responsible for plotting and saving confusion matrix of its given dataset and model.

```python
t_data = pd.read_csv('data/last_train.csv')
data = pd.read_csv('data/last_test.csv')
X_train,y_train = t_data.values[:,:-1].astype('float'), t_data.values[:,-1].astype('int')
X_test,y_test = data.values[:,:-1].astype('float'), data.values[:,-1].astype('int')

train_evaluator = evaluator.Evaluator(X_train,y_train)
test_evaluator = evaluator.Evaluator(X_test,y_test)
```

Each figure cell consists of two confusion matrices and two classification reports. The first confusion matrix and classification report belong to the training set. The second ones belong to the test set.
