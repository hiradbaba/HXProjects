```python
'''Train and Validation Dataset'''
train_mask_path_male = "./data/Train/WithMask/Male/"
train_mask_path_female = "./data/Train/WithMask/Female/"

train_wmask_path_male = "./data/Train/WithoutMask/Male/"
train_wmask_path_female = "./data/Train/WithoutMask/Female/"

train_nperson_path = "./data/Train/NotPerson/NotPerson/"


'''read data form dataset'''
with_mask_data_male = [train_mask_path_male + x for x in os.listdir(train_mask_path_male)]
with_mask_size_male = [Image.open(train_mask_path_male + x).size for x in os.listdir(train_mask_path_male)]

with_mask_data_female = [train_mask_path_female + x for x in os.listdir(train_mask_path_female)]
with_mask_size_female = [Image.open(train_mask_path_female + x).size for x in os.listdir(train_mask_path_female)]

with_mask_data = with_mask_data_male + with_mask_data_female
with_mask_size = with_mask_size_male + with_mask_size_female

without_mask_data_male = [train_wmask_path_male + x for x in os.listdir(train_wmask_path_male)]
without_mask_size_male = [Image.open(train_wmask_path_male + x).size for x in os.listdir(train_wmask_path_male)]

without_mask_data_female = [train_wmask_path_female + x for x in os.listdir(train_wmask_path_female)]
without_mask_size_female = [Image.open(train_wmask_path_female + x).size for x in os.listdir(train_wmask_path_female)]

without_mask_data = without_mask_data_male + without_mask_data_female
without_mask_size = without_mask_size_male + without_mask_size_female


not_person_data = [train_nperson_path + x for x in os.listdir(train_nperson_path)]
not_person_size = [Image.open(train_nperson_path + x).size for x in os.listdir(train_nperson_path)]


'''Train labels'''
#with mask
y_mask = [1 for x in with_mask_data]
#without mask
y_wmask = [0 for y in without_mask_data]
#not person
y_nperson = [2 for y in not_person_data]

train_df = pd.DataFrame({
    'image': with_mask_data + without_mask_data + not_person_data,
    'label': y_mask + y_wmask + y_nperson,
    'size': with_mask_size + without_mask_size + not_person_size
})


train_df.sort_values(by=['label']).to_csv('./csv/train.csv',index=False)
```

```python
'''Male and Female Test'''
test_mask_f_path = "./data/Test/WithMask/Female/"
test_mask_m_path = "./data/Test/WithMask/Male/"
test_wmask_f_path = "./data/Test/WithoutMask/Female/"
test_wmask_m_path = "./data/Test/WithoutMask/Male/"
test_not_person_path = "./data/Test/NotPerson/NotPerson/"


'''Create datasets out of the images in the FM folder'''
without_mask_f_test = [test_wmask_f_path + x for x in os.listdir(test_wmask_f_path)]
without_mask_f_test_size = [Image.open(test_wmask_f_path + x).size for x in os.listdir(test_wmask_f_path)]

without_mask_m_test = [test_wmask_m_path + x for x in os.listdir(test_wmask_m_path)]
without_mask_m_test_size = [Image.open(test_wmask_m_path + x).size for x in os.listdir(test_wmask_m_path)]

with_mask_f_test = [test_mask_f_path + x for x in os.listdir(test_mask_f_path)]
with_mask_f_test_size = [Image.open(test_mask_f_path + x).size for x in os.listdir(test_mask_f_path)]

with_mask_m_test = [test_mask_m_path + x for x in os.listdir(test_mask_m_path)]
with_mask_m_test_size = [Image.open(test_mask_m_path + x).size for x in os.listdir(test_mask_m_path)]

not_person_test = [test_not_person_path + x for x in os.listdir(test_not_person_path)]
not_person_test_size = [Image.open(test_not_person_path + x).size for x in os.listdir(test_not_person_path)]


'''Male Female Annotation'''
yt_mask_f = [1 for x in with_mask_f_test]
yt_mask_m = [1 for x in with_mask_m_test]

yt_wmask_f = [0 for x in without_mask_f_test]
yt_wmask_m = [0 for x in without_mask_m_test]

yt_not_person = [2 for x in not_person_test]


'''Cast to CSV'''
test_df_f = pd.DataFrame({
    'image': with_mask_f_test + without_mask_f_test,
    'label': yt_mask_f + yt_wmask_f,
    'size' : with_mask_f_test_size + without_mask_f_test_size
})

test_df_m = pd.DataFrame({
    'image': with_mask_m_test + without_mask_m_test,
    'label': yt_mask_m + yt_wmask_m,
    'size' : with_mask_m_test_size + without_mask_m_test_size
})

test_df_fm = pd.DataFrame({
    'image': with_mask_m_test + without_mask_m_test + with_mask_f_test + without_mask_f_test,
    'label': yt_mask_m + yt_wmask_m + yt_mask_f + yt_wmask_f,
    'size' : with_mask_m_test_size + without_mask_m_test_size +  with_mask_f_test_size + without_mask_f_test_size,
})

test_df_fmn = pd.DataFrame({
    'image': with_mask_m_test + without_mask_m_test + with_mask_f_test + without_mask_f_test + not_person_test,
    'label': yt_mask_m + yt_wmask_m + yt_mask_f + yt_wmask_f + yt_not_person,
    'size' : with_mask_m_test_size + without_mask_m_test_size +  with_mask_f_test_size + without_mask_f_test_size + not_person_test_size,
})

'''Serialize'''
test_df_f.sort_values(by=['label']).to_csv('./csv/test_f.csv',index=False)
test_df_m.sort_values(by=['label']).to_csv('./csv/test_m.csv',index=False)
test_df_fm.sort_values(by=['label']).to_csv('./csv/test_fm.csv',index=False)
test_df_fmn.sort_values(by=['label']).to_csv('./csv/test_fmn.csv',index=False)
```
