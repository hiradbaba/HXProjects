```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
import math
df = pandas.read_csv('files/train.csv')
```

```python
def calPriceError():
    valid = pandas.read_csv('files/validation.csv')
    predict = valid['Predicted']
    actual = valid['Actual']
    rmsle = valid['RMSLE']
    summ = 0
    ss = 0
    count = 0
    for p,a in zip(predict,actual):
        summ+= a-p
        ss+= (a-p)**2
        count += 1
    print("The avg difference in price is:{}".format(summ//count))
    print("the rmse is: {}".format(int(math.sqrt(ss/count))))
    print("the rmsle is: {}".format(rmsle[0]))
    
def plotError():
    plt.figure()
    ef = pandas.read_csv('files/validation.csv')
    print("average accuarcy rmsle: {}".format(ef['Mean_accuarcy'][0]))
    ef.plot.line(x='id',y='Accurcy',color='#ff7700')
    calPriceError()
```
